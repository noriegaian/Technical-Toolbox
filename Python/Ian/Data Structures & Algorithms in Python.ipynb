#### Tuples vs Lists vs Sets vs Dictionaries (PSADS):

- **List**: An ordered collection of zero or more references to Python data objects
         Written in [] and can be heterogenous
         Ex: [1,3,True,"Ian"]
- **Tuple**: Similar to lists except immutable (cannot be changed)
         Written with ()
         Ex: (1,3,True,"Ian")
- **Set**: An unordered collection of zero or more immutable Python data objects
         Written with {}
         An empty set is represented by set() (not to be confused with dictionary)
         Sets do not allow duplicates
         Ex: {1,3,True,"Ian"}
- **Dictionary**: An unordered collection of key-value pairs
         Written as key:value pairs in {}
         Ex: {'Name': 'Ian', 'Country': 'Canada'}

#### Monkey Trouble (CodingBat)

There are two monkeys and the parameters indicate if each is smiling. We are in trouble if they are either both smiling or neither are smiling. Return true if we are in trouble.

Tags: boolean, if else

```python
a_smile = True
b_smile = False

def monkey_trouble(a_smile, b_smile):
    if a_smile and b_smile: # note that for booleans we don't have to specify == True
        return True
    elif not a_smile and not b_smile: # ^^ or == False (use not)
        return True
    else:
        return False

monkey_trouble(a_smile, b_smile)
```

#### Rock Paper Scissors (Code Wars)

If two players are playing rock-paper-scissors, write a function to return which player wins. Return 'Draw!' if it's a tie.

Tags: hashmap, dictionary, if else

```python
# note that I initially used if else logic here but hashmap is so much cleaner!

p1 = 'scissors'
p2 = 'paper'

def rps(p1, p2):
    # create a dictionary to use as our rulebook (key is winning move)
    rulebook = {'scissors': 'paper', 'paper': 'rock', 'rock': 'scissors'}
    if rulebook[p1] == p2:
        return 'Player 1 wins!'
    elif rulebook[p2] == p1:
        return 'Player 2 wins!'
    else:
        return 'Draw!'

rps(p1, p2)
```

#### Fizz Buzz (LeetCode)

Given an integer n, return a STRING array that is 1-indexed (dumb) where:
- "FizzBuzz" if divisible by 3 and 5
- "Fizz" if divisible by 3
- "Buzz" if divisible by 5
- The number otherwise

Tags: if else, elif, for loop, order of operations

```python
n = 6

def fizzbuzz(n):
    arr = []
    for x in range(1,n+1):
        if x % 5 == 0 and x % 3 == 0:
            value = 'FizzBuzz'
            arr.append(value)
        elif x % 5 == 0:
            value = 'Buzz'
            arr.append(value)
        elif x % 3 == 0:
            value = 'Fizz'
            arr.append(value)
        else:
            value = x
            arr.append(value)
    return arr

fizzbuzz(n)
```

#### Detecting an Anagram (PSADS)

One string is an anagram of another if the second is just a rearrangement of the characters in the first. Write a function that takes two strings and returns whether or not they are anagrams. We'll assume that both strings are of equal length.

Tags: string manipulation, loop, string to list, sort

```python
# first solution: "sort and compare"
    # sort the characters in each string alphabetically and then compare

s1 = 'python'
s2 = 'typhon'

def anagram_solution_1(s1, s2):
    l1 = sorted(s1) # converts to a sorted list (alphabetically)
    l2 = sorted(s2)
    
    if l1 == l2:
        return True
    else:
        return False

anagram_solution_1(s1, s2)

# Complexity: O(n^2) or O(n log n) typically when you sort
```

Tags: complexity, big O formula explanation, ord, unicode, alphabet

```python
# second (better) solution: "count and compare"
    # uses the fact that anagrams will have the same # of each letter (e.g., a, b, c, etc.)
    # if counts for all 26 letters

s1 = 'python'
s2 = 'typhon'
    
def anagram_solution_2(s1, s2):
    c1 = [0]*26 # create empty arrays to store letter counts
    c2 = [0]*26
    
    for i in range(len(s1)):
        pos = ord(s1[i]) - ord('a') # ord can be used to return unicode code for a character
        c1[pos] += 1 # each time position occurs, we increment the count
        
    for i in range(len(s2)):
        pos = ord(s2[i]) - ord('a')
        c2[pos] += 1
        
    k = 0
    anagramYes = True
    while k < 26 and anagramYes: # while loop for as long as we're in limits of the alphabet
        if c1[k] == c2[k]: # now we're checking each character count
            k += 1 # if they match we increment position in our arrays
        else:
            anagramYes = False
    
    return anagramYes

anagram_solution_2(s1, s2)

# Complexity: O(n) as both character counts are based on n and then there are
    # 26 steps (c) for comparing the counts...therefore T(n) = 2n + 26
```

#### Palindrome (LeetCode)

Given a string, write a python function to check if it is a palindrome or not (i.e., same word backwards).

Tags: string manipulation, reversing, sort, two pointer, for loop

```python
# first solution: "reverse and compare"

word = 'anna'

def is_palindrome(word):
    word_rev = word[::-1] # reverse the string
    
    if word_rev == word:
        return True
    return False

is_palindrome(word)

# Complexity: O(n) - goes through entire string once
```

```python
# second solution: "two pointer"
    # idea here is to "point" at two elements at the same time
    # iff palindrome, first and last elements should be the same, progressively move inwards

word = 'africa'

def is_palindrome(word):
    pointer1 = 0
    pointer2 = len(word)-1
    while pointer1 < pointer2:
        # if elements are not equal return false
        if word[pointer1] != word[pointer2]:
            return False
        else:
        # if elements are equal, compare next ones
            pointer1 += 1
            pointer2 -= 1
    return True
    
is_palindrome(word)

# Complexity: Still O(n) but would be faster if the string is not a palindrome
```

#### Last Two (CodingBat)

Given a string, return the count of the number of times that a substring length 2 appears in the string and also as the last 2 chars of the string.

Tags: string manipulation, for loop, substring

```python
s = 'hixxhihi'

def last2(s):
    # first cover the case if string is less than 2 char:
    if len(s) < 2:
        return 0
    else:
        last_two = s[-2:] # identify last 2 char
        count = 0 # initialize count
        
        # we only want to set for loop up until last 2 chars
        for i in range(len(s)-2):
            substr = s[i:i+2] # cycle through 2L substrings
            if substr == last_two:
                count += 1
        return count

last2(s)
```

#### String Match (CodingBat)

Given 2 strings, a and b, return the number of positions where they contain the same length 2 substring.

Tags: string manipulation, for loop, substring

```python
a = 'xxcaazz'
b = 'xxbaaz'

def string_match(a,b):
    count = 0 # initialize count
    min_len = min(len(a), len(b)) # determine length of shorter string
    
    #we only need to find all 2L substrings that exist in the shorter string
        # then we check to see if they exist in the longer string
    for i in range(0,min_len-1):
        sub_a = a[i:i+2]
        sub_b = b[i:i+2]
        if sub_a == sub_b:
            count += 1
    return count

string_match(a,b)
```

#### Do Elements in List Satisfy some Constraint (ATDSI)

Write a function to determine if a list contains two elements that sum up to some value. This solution makes use of hashmaps (same as dictionary in python).

Tags: hashmap, dictionary, complement, sum

```python
arr = [3,1,4,2,6,9]
target = 11

def check_sum(arr, target):
    hashmap = {} # create empty hashmap (dictionary)
    for i, e in enumerate(arr):
        if (target - e) in hashmap: # check hashmap
            return True
        else:
            hashmap[e] = e # add to hashmap
    return False

check_sum(arr, target)
```

#### Two Sum (LeetCode)

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to target. Cannot use the same element twice.

Tags: hashmap, dictionary, complement, sum

```python
nums = [2,7,11,15]
target = 17

# Note that a hashmap is the best way to maintain a mapping of each element 
    # in the array to its index

def twosum(nums, target):
    hashmap = {} # create empty hashmap (dictionary)
    for i in range(len(nums)): # we want to search by index hence range(len)
        complement = target - nums[i] # defining the complement
        if complement in hashmap:
            return [i, hashmap[complement]]
        hashmap[nums[i]] = i
        
twosum(nums,target)
    
# this is O(n) complexity
```

#### A) Linear Search (Jovian)

##### Comprehensive approach to problem

You have some cards with numbers written on them. Cards are arranged in decreasing order and laid face down on a table. You want to pick out the card containing a given number by turning over as few cards as possible.



Tags: linear search, test cases, while loop

A few important cases we should brainstorm and consider **ahead of coding** (includes edge cases):

    2) The number is the first element

Now that we have our test cases, let's try to propose a solution in english / pseudocode:

- Create a variable `position` with the value 0
- Check whether the number at index `position` is equal to `query`
- If it does, `position` is the answer and should be returned
- If not, increment the value of `position` by 1, and repeat steps 2-5 until we reach the final position
- If there is still no match, return -1 (recall our edge cases)
    
This is the brute force solution (simplest, most obvious solution to a problem, may not be the most optimal). More specifically, it's a **linear search** algorithm as we're searching through the list in linear fashion element-by-element. **Linear search is O(n) complexity.**

```python
# let's write our function now

    # testing out case 6
cards = []
query = 0

def locate_card(cards, query):
    position = 0 # create variable
    while position < len(cards): # set up while loop for repetition
        if cards[position] == query: # check if element at current position matches query
            return position
        position += 1 # increment the position
    return -1

locate_card(cards, query)

# this is O(n) complexity (at most we go through the numbers array once)
```

#### B) Binary Search (Jovian)

##### Same problem as above but more efficient solution

You have some cards with numbers written on them. Cards are arranged in decreasing order and laid face down on a table. You want to pick out the card containing a given number by turning over **as few cards as possible**.



Tags: binary search, test cases, while loop

An **important** realization we didn't account for in our linear search solution was that the cards are already arranged in decreasing order. We simply constructed a brute force approach where we start by picking the first card.

Realistically, we can pick the first card from the **middle** (e.g., pick a 7 in the middle) and we would know that a) if we're looking for a query > 7 the card must be to the left or b) if we're looking for a query < 7 the card must be to the right. This quickly eliminates multiple possibilities...keep repeating this until you systematically get to the query card.

Again, let's state our new and improved approach in plain english:

- Create variable `position` which is in the middle of the list
    - What if there is an even number of cards?
- Check whether the number at index `position` is equal to `query`
- If it does, `position` is the answer and should be returned
- If not, check if the number at index `position` is > or < to `query`
    - If greater, change `position` to the middle of the list end and last `position`
    - If smaller, change `position` to the middle of the list start and last `position`
- Repeat steps until we reach the final position
- If there is still no match, return -1 (recall our edge cases)

```python
cards = [8, 7, 4, 2, -1]
query = -1

def locate_card(cards, query):
    low, high = 0, len(cards) - 1 # define low and high bounds
    while low <= high: # while loop (low bound should never exceed high)
        #obtain midpoint (// to handle even # of cards)
        position = (low + high) // 2
        
        # add an intermediate check to make sure numbers are working
        print('low:', low, "high:", high, "position:", position)
        
        if cards[position] == query:
            return position
        elif cards[position] > query:
            low = position + 1
            # we essentially shift the lower bound just right of our midpoint
        elif cards[position] < query:
            high = position - 1
            # similarly we shift the high bound just left of midpoint
    
    return -1

locate_card(cards, query)

# this is O(log n) complexity
```

Now we have our **binary search** algorithm. This differs from our linear search approach (searches list element-by-element linearly), as with binary search we are not accessing elements in any order, but rather in a pseudo-random fashion (taking the midpoint repetitively). Important to note that binary search requires a **sorted** list/array. **Binary search is O(log n) complexity**.

#### Sort the Odds (Code Wars)

Given an array of numbers, sort the odd numbers in ascending order while leaving the even numbers at their original positions.

Tags: array, sort, insert, list comprehension, enumerate

```python
numbers = [5,8,6,3,4]

def sort_array(numbers):
    result = sorted([x for x in numbers if x%2 != 0]) # return a sorted array of odd numbers using list comprehension
    for i, x in enumerate(numbers): # enumerate allows us to iterate through a sequence but keeps track of index and element
        if x % 2 == 0: # iff even
            result.insert(i,x) # insert the even number at required index
    return result

sort_array(numbers)
```

#### Move Zeroes (LeetCode)

Given an integer array, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Tags: array, sort, order, reorder, for loop, while loop

```python
nums = [0,1,0,3,12]

def moveZeroes(nums):
    count = 0 # initialize count of non-zero elements
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[count] = nums[i] # replacing the element at index with the non-zero element (essentially shifting these elements to the left)
            count += 1 # incrementally increase count (will eventually be index of 1st zero)
            
    while count < len(nums): # now for the zero elements
        nums[count] = 0
        count += 1
        
    return nums

moveZeroes(nums)
```

#### Intersection of Two Arrays (ATDSI)

Given two arrays, write a function to get the intersection for the two.

Tags: arrays, intersection, sets, list comprehension, in

```python
arr1 = [1,2,3,4,5]
arr2 = [0,1,3,7,9,10]

def findintersection(arr1,arr2):
    set1 = set(arr1) # turn into a set (no duplicates)
    set2 = set(arr2)
    if len(set1) > len(set2):
        return [x for x in set1 if x in set2] # we want to return an array
    elif len(set2) > len(set1):
        return [x for x in set2 if x in set1] # use list comprehension

findintersection(arr1,arr2)
```

#### Max Product of 3 Numbers in Array (ATDSI)

Given an integer array, return the maximum product of any three numbers in the array.

```python
# brute force - checking every subsequent triplet after sorting

arr = [-2,-4,5,3]
arr2 = [1,-10,-5,3,10,9,1]

def MaxTriProduct(arr):
    ordered = sorted(arr) # first sort the array
    results = []
    for i in range(len(ordered)):
        a = ordered[i]
        b = ordered[i-1] # important to go negative
        c = ordered[i-2]
        product = a*b*c
        results.append(product)
    return max(results)

MaxTriProduct(arr)

# Complexity: O(n log n) due to sorting
```

#### Largest Contiguous Sum of Subarray (ATDSI)

Given an integer array, find the sum of the largest contiguous subarray within the array. Note that if all elements are negative, you should return 0.

Tags: for loop, continuous, subsequent, array

```python
# tricky one
arr = [-1,-3,5,-4,3,-6,9,-1,10,1]

def LargestContSum(arr):
    max_sum = 0 # initialize max sum
    current_sum = 0 # initialize current sum
    for x in arr:
        current_sum = max(0, current_sum + x) # will reset to 0 if it goes into negatives
        max_sum = max(max_sum, current_sum) # get the max
    return max_sum
        
LargestContSum(arr)

# Complexity: O(n)
```

```python
nums = [1,1,0,1,1,1,1]

def consec_ones(nums):
    current_count = 0
    max_count = 0 # initializing counts
    
    for i, e in enumerate(nums): #simply getting index and element of array
        if e == 1:
            current_count += 1
            max_count = max(max_count, current_count)
        elif e != 1:
            current_count = 0 # reset once we run into a non-1
    return max_count
                
consec_ones(nums)
```

```python
# one option (somewhat confusing leetcode?)
nums = [0,0,1,1,1,2,2,3,3,4]

def removeDuplicates(nums):
    inc = 1 # initialize the index we'll add new #'s to in resulting array
    # we're going to compare each element to the one after it
    for i in range(len(nums)-1):
        if nums[i] != nums[i+1]:
            nums[inc] = nums[i+1] # assign the diff number to subsequent index
            inc += 1
    return nums

removeDuplicates(nums)

# Complexity: O(n)
```

```python
# optimal answer - two pointer strat
nums = [-7,-1,0,3,10]

def sortedSquares(nums):
    result = [0] * len(nums) # empty array w/ same length
    pointer1 = 0 # initializing pointers
    pointer2 = len(nums) - 1
    
    # we know w/ sorted array, either the 1st or last element will be largest square
    for i in range(len(nums)-1,-1,-1): # set the for loop backwards
        if abs(nums[pointer1]) > abs(nums[pointer2]): # if negative > positive
            result[i] = nums[pointer1] ** 2 # insert at back of our array
            pointer1 += 1 # incrementally push pointer 1 up
        else:
            result[i] = nums[pointer2] ** 2
            pointer2 -= 1 # incrementally push pointer 2 down
    return result
    
sortedSquares(nums)
```

#### Normally Distributed List Check

Source: https://priyankabanerjee.medium.com/guide-for-data-science-interview-python-71901df4bd44

Write a function that checks if a list of integers is normally distributed. Specifically, write a function that returns a score that measures the deviation from normality.

Recall standard deviation formula:


Tags: stats, formula, for loop, if else

```python
import numpy as np
import random

# generate a normally distributed sample to test on
sample_list = np.random.normal(50,15,10000).round()

sample_list #preview it
```

```python
# calculate mean and standard deviation
def normality_test(sample_list):
    length = len(sample_list)
    mean = sum(sample_list)/length
    stdev = (sum((x - mean)**2 for x in sample_list) / length) ** 0.5
# calculate % of values within each SD via empirical rule (68/95/99.7)
    num1, num2, num3 = 0, 0, 0
    for x in sample_list:
        if (x < mean + stdev) and (x > mean - stdev):
            num1 += 1
        if (x < mean + 2*stdev) and (x > mean - 2*stdev):
            num2 += 1
        if (x < mean + 3*stdev) and (x > mean - 3*stdev):
            num3 += 1
    d1 = num1 / length
    d2 = num2 / length
    d3 = num3 / length
    return [d1,d2,d3]

normality_test(sample_list)
```
